```python
inmuebles['Valor_m2_Millon'] = inmuebles['Precio_Millon']/inmuebles['Area'] #Para cada fila divide el precio de millon en area de cada inmueble
inmuebles.head()
```

```python
inmuebles.groupby('Barrio').mean()
```

```python
datos_barrio = inmuebles.groupby('Barrio').sum()
datos_barrio
```

```python
datos_barrio['Valor_m2_Barrio'] = datos_barrio['Precio_Millon']/datos_barrio['Area']
datos_barrio
```

```python
m2_barrio = dict(datos_barrio['Valor_m2_Barrio'])
```

```python
inmuebles['Valor_m2_Barrio'] = inmuebles['Barrio']
inmuebles['Valor_m2_Barrio'] = inmuebles['Valor_m2_Barrio'].map(m2_barrio) #Cambia los barrios por el valor que tiene cada uno en el dictionary
inmuebles.head()
```

```python
top_barrios = inmuebles['Barrio'].value_counts()[:10].index   #En panda series, el indice es el primero y sigue valor
```

```python
datos_barrio.reset_index(inplace=True) #resetea el indice
datos_barrio
```

```python
datos_barrio.query('Barrio in @top_barrios') #Se usa la @Para referirse a variable
```

```python
plt.figure(figsize=(10,8))
ax = sns.barplot(x="Barrio", y= "Valor_m2_Barrio", data=datos_barrio.query('Barrio in @top_barrios'))
ax.tick_params(axis='x', rotation=45) #Los tick son donde estan cada punto que queremos indicar
```

```python
plt.figure(figsize=(10,8))
ax = sns.boxplot(x="Barrio", y= "Valor_m2_Millon", data = inmuebles.query('Barrio in @top_barrios'))
ax.tick_params(axis='x', rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,8))
ax = sns.boxplot(x="Barrio", y= "Valor_m2_Millon", data = inmuebles.query('Barrio in @top_barrios & Valor_m2_Millon < 12'))
ax.tick_params(axis='x', rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,8))
ax = sns.boxplot(x="Barrio", y= "Area", data = inmuebles.query('Barrio in @top_barrios & Area < 400'))
ax.tick_params(axis='x', rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,8))
ax = sns.boxplot(x="Barrio", y= "Precio_Millon", data = inmuebles.query('Barrio in @top_barrios & Precio_Millon < 2000'))
ax.tick_params(axis='x', rotation=45)
plt.show()
```
