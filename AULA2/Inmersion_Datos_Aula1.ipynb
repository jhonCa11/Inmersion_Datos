```python
valor = inmuebles.Valor.str.split(expand = True) #El expand=True, devuelve el resultado en formato de DataFrame (No panda series)
inmuebles['Moneda'] = valor[0] #Genera una nueva columna, y le asigna un valor
inmuebles['Precio'] = valor[1]
inmuebles.sample(3)
```

```python
inmuebles.info()
```

```python
inmuebles['Precio'] = inmuebles['Precio'].str.replace('.', '',regex=True)
```

```python
inmuebles[['Precio', 'Barrio']] #Colocarlo como en un formato de lista, para que se entienda que se desea un datFrame
```

```python
inmuebles['Precio_Millon'] = inmuebles.Precio.astype('float')/1000000   #El astype, cambia el tipo de un dato - se divide para poder visualizar mejor el número
inmuebles.info()
```

```python
inmuebles.describe()  #Describe todos los datos
```

```python
pd.set_option('display.precision', 2) #set_option(), función de pandas - display.precision, parametro de precisión con número
pd.set_option('display.float_format', lambda x: '%.2f' %x) #formato de datos, con 2 números despues del.
inmuebles.describe()
```

```python
inmuebles.loc[inmuebles.Habitaciones == 110]  #El .loc[], permite localizar un atributo que cumpla la condición
```

```python
inmuebles.loc[inmuebles.Area == 2]
```

```python
inmuebles['Precio_Millon'].plot.hist(bins = 100) #histograma(bins, rango que abarca en el hist)
```

```python
#Importar paquetes
import matplotlib.pyplot as plt
import seaborn as sns

#Preparar lienzo, para colocar grafica
plt.figure(figsize=(10,7)) #Crea el lienzo de ancho 10, largo 8
grafica = sns.histplot(data=inmuebles, x='Precio_Millon', kde=True, hue='Tipo') #histograma con seaborn - data, de donde los datos -kde: Curva de densidad - hue() añade otra dimensión a nuestro grafico
grafica.set_title('Distribución de Valores de inmuebles Bogotá') #Titulo
plt.xlim((50, 1000))  #limite en eje X (min, max)
# plt.ylim((0,100)) #Limita en Y (Zoom)

plt.savefig('/content/drive/MyDrive/inmersion_datos/valor_inmuebles.png',format='png') #savefig('ruta/nombrearchivo.extension', extension), guarda grafica
plt.show()
```
